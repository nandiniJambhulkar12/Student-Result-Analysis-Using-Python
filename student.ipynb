```python
import numpy as np   
import pandas as pd   
import matplotlib.pyplot as plt   
import seaborn as sns
```

```python
df=pd.read_csv("Expanded_data_with_more_features.csv")
print(df.head())
```

```python

# drop unamed column
```

```python
df = df.drop("Unnamed: 0", axis = 1  )
print(df.head())
```

```python
# wkly study hrs change
```

```python
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("5 - 10 ","5 - 10 ")
df.head()
```

```python
# gender distribution
```

```python
plt.figure(figsize=(5,5))
ax=sns.countplot(data=df,x="Gender")
ax.bar_label(ax.containers[0])
plt.show
```

```python
# from the above chart we have analyse that the number of females in the data is more than the number of males.
```

```python
gb=df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":"mean"})
print(gb)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb, annot = True) 
plt.title("Gender Distributuion")
plt.show()
```

```python
# from the above chart we have concluded that the education of the parents have a good impact on their scores.
```

```python
gb1=df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":"mean"})
print(gb1)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb1, annot = True) 
plt.title("Relationship between parents education and student score")
plt.show()
```

```python
# from the above chart we have concluded that rhere is no impacton the student scores due toperents maretal status.
```
